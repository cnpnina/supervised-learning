```python
# Ex1
import matplotlib.pyplot as plt
def cal_fx(x): #cost function
  fx = x**2 - 6*x + 16
  return fx


def cal_grad(x):
  grad_fx = 2*x - 6
  return grad_fx


# initial value
x = 5
n_iter = 5500 #ค่า batch size
learning_rate = 0.002 #alpha
X = []
iter = []

for i in range(n_iter):
  x = x - learning_rate * cal_grad(x)
  fx = cal_fx(x)
  X.append(x)
  iter.append(i)

print('Result of x = %s' % x)
print('Result of f(x) = %s' % fx)
plt.xlabel('number of iterations')
plt.ylabel('x')
plt.plot(iter , X, linewidth=1)
plt.show()
```

```python
# Ex2
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error
data = pd.read_csv('Salary_Data.csv')
X = data['YearsExperience'].to_numpy().reshape(-1,1)
Y = data['Salary'].to_numpy().reshape(-1,1)

def cal_grad(X, y, theta):
 m = len(X)
 grad_theta = np.array([0, 0])
 for i in range(m):
  grad_theta[0] += ((theta[0] + theta[1]*X[i]) - y[i])
  grad_theta[1] += ((theta[0] + theta[1]*X[i]) - y[i]) * X[i]
 return grad_theta /m

def cal_fx(X, y, theta):
 m = len(X)
 J = 0
 for i in range(m):
  J += (theta[0] + theta[1]*X[i] - y[i])**2
 return J / (2*(m))

learning_rate = 0.002 #alpha
J = []
iter = []
theta = np.array([0, -1])
n_iter = 20000

for i in range(n_iter):
 theta_grad = cal_grad(X, Y, theta)
 theta = theta - learning_rate * theta_grad
 J.append(cal_fx(X, Y, theta))
 iter.append(i)

input_X = np.c_[np.ones((X.shape[0],1)), X.reshape((len(X), 1))]
y_pre = input_X.dot(theta)
print('final theta = %s' % theta)
print('final cost = %s' % J[-1])
print("Mean squared error1: %.7f" % mean_squared_error( Y, y_pre))
plt.xlabel('number of iterations')
plt.ylabel('x')
plt.plot(iter , J, linewidth=1)
plt.show()
```
