```python
n = x.size

upper_eq = sum(x * y) - sum(x)*sum(y) / n
lower_eq = sum(x * x) - n * np.mean(x) **2
m = upper_eq/lower_eq
b = np.mean(y) - m * np.mean(x)
y_pre = m * x + b
print("Mean squared error 1: %.7f" % mean_squared_error( y, y_pre))
print(m)
print(b)
```

```python
plt.plot(x,y_pre, color = 'red')
plt.scatter(x,y)
plt.title('YearsExperience with Salary Prediction')
plt.xlabel('YearsExperience')
plt.ylabel('Salary')
plt.show()
```

```python
# Ex2

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error

data = pd.read_csv('Salary_Data.csv')
x = data['YearsExperience'].to_numpy().reshape(-1,1)
y = data['Salary'].to_numpy().reshape(-1,1)

matrix_1 = []
for i in range(len(data)):
  matrix_1.append(1)
#print(matrix_1)

X = np.transpose(np.array([matrix_1, data['YearsExperience'].to_list()]))
#print(X.shape)

transposed_matrix = np.transpose(X)
#print(transposed_matrix.shape)


theta = np.matmul(np.matmul(np.linalg.inv(np.matmul(transposed_matrix, X)), transposed_matrix), y)
print(f"theta\n" ,theta)

y_pre = X.dot(theta)
print("Mean squared error 2: %.7f" % mean_squared_error(y, y_pre))
```

```python
plt.scatter(x, y)
plt.plot(x, y_pre, color = 'orange')
plt.title("YearsExperience with Salary Prediction")
plt.xlabel('YearsExperience')
plt.ylabel('Salary')
plt.show()
```

```python
# Ex3
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

data = pd.read_csv('Salary_Data.csv')
x = np.array(data['YearsExperience'])
y = np.array(data['Salary'])
X = x.reshape(-1,1)
Y = y.reshape(-1,1)

model = LinearRegression()
model.fit(X, Y)
print(model.intercept_ , model.coef_)
y_pre = model.predict(X)
print("Mean squared error 3 : %.3f" % mean_squared_error( y_pre, Y))
plt.plot(X, y_pre)
plt.scatter(X,Y, color = 'violet')
```

```python
# Ex4
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import train_test_split

data_sales = pd.read_csv('Sales.csv')


print(data_sales.info())
```

```python
print(data_sales.describe())
```

```python
import seaborn as sns
pairplot = sns.pairplot(data_sales)
```

```python
print(data_sales.corr())
dataplot = sns.heatmap(data_sales.corr(), cmap="YlGnBu", annot=True)
```

```python
#TV
X = np.array(data_sales['TV']).reshape(-1,1)
Y = np.array(data_sales['Sales']).reshape(-1,1)
X_train, X_test, y_train, y_test = train_test_split(X, Y, train_size=0.7 ,test_size = 0.3, random_state=2)
model = LinearRegression()
model.fit(X_train, y_train)
print(model.intercept_ , model.coef_)
y_pre = model.predict(X_test)

print("Mean squared error: %.3f" % mean_squared_error(y_test, y_pre))
print('r_square_value :',r2_score(y_test, y_pre))
plt.plot(X_test,y_pre)
```
